```python
# 3. Blank Lines

# A. Top-Level Functions & Classes
# Use 2 blank lines before top-level functions or class definitions.
# Helps visually separate major sections of code.

# Correct
class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model


def calculate_speed(distance, time):
    return distance / time
# There are 2 blank lines before def calculate_speed (top-level function). ✅



# B. Methods Inside Classes
# Use 1 blank line between methods in a class.

class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def start_engine(self):
        print("Engine started")

    def stop_engine(self):
        print("Engine stopped")
# Notice 1 blank line between start_engine and stop_engine. ✅



# 4. Whitespace Rules

# A. Inside Parentheses, Brackets, or Braces
# No spaces immediately inside: (), [], {}

# Correct ✅
my_list = [1, 2, 3]
result = func(a, b)

# Incorrect ❌
my_list = [ 1, 2, 3 ]
result = func( a, b )


# B. Around Operators
# One space on both sides of binary operators (=, +, -, *, /, ==, <, >, etc.)

# Correct ✅
x = 5
total = a + b - c

# Incorrect ❌
x=5
total=a+b-c

# No spaces around = when used for keyword arguments or default values:
def greet(name="World"):  # Correct ✅
    print(f"Hello, {name}")
```
